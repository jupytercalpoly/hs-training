```python
import pandas as pd
```

```python
cases = pd.read_csv('us_confirmed.csv', index_col=0)
```

```python
cases.head(10)
```

```python
subset = cases.copy().iloc[5:,]
```

```python
subset = subset[['UID','Admin2','Province_State','6/1/20']].rename(columns={'6/1/20':'cases'})
```

```python
deaths = pd.read_csv('us_deaths.csv', index_col=0)
```

```python
subset_d = deaths.copy().iloc[5:,]
```

```python
subset_d = subset_d[['UID','6/1/20']].rename(columns={'6/1/20':'deaths'})
```

```python
df = subset.merge(subset_d, on='UID')
```

```python
df.head()
```

```python
df["FIPS"] = df['UID'].astype(str).str[-5:].astype(str)
```

```python
df["FIPS"]
```

```python
df.to_csv('counties_june1.csv')
```
